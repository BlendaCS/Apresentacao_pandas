```python
df["Custo Total"] = df["Custo Unitário"].mul(df["Quantidade"])
```

```python
df.head(5)
```

```python
df.rename(columns={"Custo Total":"Custo"}, inplace = True)
```

```python
df.head(1)
```

```python
round(df["Custo"].sum(),2)
```

```python
df["Lucro"] = df["Valor Venda"] - df["Custo"]
```

```python
df.head(1)
```

```python
round(df["Lucro"].sum(),2)
```

```python
df["Tempo_envio"] = df["Data Envio"] - df["Data Venda"]
```

```python
df.head(1)
```

```python
df["Tempo_envio"] = (df["Data Envio"] - df["Data Venda"]).dt.days
```

```python
df.head(1)
```

```python
df.dtypes
```

```python
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=False) 
```

```python
df.groupby("Produto",)["Quantidade"].sum().sort_values(ascending=True).plot.barh(title = "Total de Produtos Vendidos")
plt.xlabel("Total")
plt.ylabel("Produtos");
```

```python
df.groupby(df["Data Venda"].dt.year)["Lucro"].sum().plot.bar(title="Lucro em cada ano")
plt.xlabel("Ano")
plt.ylabel("Lucro");
```

```python
df.groupby(df["Data Venda"].dt.year)["Lucro"].sum()
```

```python
df_2009 = df[df["Data Venda"].dt.year == 2009]
```

```python
df_2009.head()
```

```python
df_2009.groupby(df_2009["Data Venda"].dt.month)["Lucro"].sum().plot(title= "Lucro por mês no ano de 2009", marker="o")
plt.xlabel("Mês")
plt.ylabel("Lucro");
```

```python
df_2009.groupby("Marca")["Lucro"].sum().plot.bar(title= "Lucro por Marca")
plt.xlabel("Marcas")
plt.ylabel("Lucro");
plt.xticks(rotation="horizontal")
```

```python
df_2009.groupby("Classe")["Lucro"].sum().plot.bar(title= "Lucro por Classe")
plt.xlabel("Classe")
plt.ylabel("Lucro");
plt.xticks(rotation="horizontal")
```
